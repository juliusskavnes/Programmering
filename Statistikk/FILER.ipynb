```python
from pylab import *
data = loadtxt('data.csv',skiprows=1,delimiter=",")

tid = data[:,0]
akselerasjon = data[:,3]

plot(tid,akselerasjon,color='forestgreen')
xlabel('Tid')
ylabel('Akselerasjon')
grid()
show()

snitt = mean(akselerasjon)
avvik = std(akselerasjon)

print("Akselerasjonen var på:", round(snitt,2), "+-", round(avvik,2))

a, b, c = polyfit(tid, akselerasjon, 2)

def modell(x):
    return a*x**2 + b*x + c

x = linspace(0,50,10)
y = modell(x)
plot(x,y,label='Tilpasset kurve')
plot(tid,akselerasjon,linestyle='',marker='.',label='Datapunkter')
legend()
show()
```
